```python
import numpy as np
import matplotlib.pyplot as plt

t= np.arange(0., 5., 0.2)
t2= t **2
t3= t**3;
plt.plot(t,t2, label="t2")
plt.plot(t,t3, label="t3")
plt.title("ok")
plt.legend()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(t,t2, label="t2")
plt.legend()
plt.title("apple")
plt.subplot(1,2,2)
plt.plot(t,t3, label="t3")
plt.legend()
plt.title("pear")
plt.show()
```
